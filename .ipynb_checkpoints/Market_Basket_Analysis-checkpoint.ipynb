```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sqlalchemy import create_engine
```

#### Connecting to Database.

```python
def db_connect():
    username = 'postgres'
    password = ''
    host = 'localhost'
    port = 5432  # Default PostgreSQL port is 5432
    database = 'postgres'
    # Create connection string and engine
    connection_string = f'postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}'
    engine = create_engine(connection_string)

    # Query Exection.
    query = 'SELECT * FROM CUSTOMER;'

    return engine,query
```

```python
# Call the function to get the engine and query
engine, query = db_connect()
```

```python
df = pd.read_sql(query, engine)
# Display the DataFrame
df.head()
```

```python
def wrangle():
    # Reading the dataset.
    df = pd.read_csv('cars_dataset.csv')
    # Dropping the unwanted columns
    df.drop(columns=['Unnamed: 0'],inplace=True)
    
    
    return df
```

```python
df = wrangle()
```

```python
df
```

```python
# Selecting the specific column from the database.
X = df.drop(columns=['customer_id','first_name','last_name','phone_num'])
```

```python
X
```

```python
transactions = []
for i in range(0, 19):
    transactions.append([str(X.values[i,j]) for j in range(0, 3)])

# Training the Apriori model on the dataset
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)

## Displaying the first results coming directly from the output of the apriori function
results = list(rules)
results

## Putting the results well organised into a Pandas DataFrame
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
resultsinDataFrame
```

```python
## Displaying the results sorted by descending lifts
resultsinDataFrame.nlargest(n = 43, columns = 'Lift')
```
